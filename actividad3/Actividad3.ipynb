```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mpl
```

```python
dfj = pd.read_csv('inuvik.txt', skiprows=5, sep='\s+', header=0)
dfd = pd.read_csv('inuvik2.txt', skiprows=5, sep='\s+', header=0)
dfj.columns = ['PRES','HGHT','TEMP','DWPT','RELH','MIXR','DRCT','SKNT','THTA','THTE','THTV']
dfd.columns = ['PRES','HGHT','TEMP','DWPT','RELH','MIXR','DRCT','SKNT','THTA','THTE','THTV']
```

```python
# Eliminamos datos incompletos
dfj = dfj.drop(dfj.index[len(dfj)-2])
dfd = dfd.drop(dfd.index[len(dfd)-1])
```

```python
# Dar estructura de datos (DataFrame)
dfj= pd.DataFrame(dfj)
dfd = pd.DataFrame(dfd)
```

```python
#GrÃ¡fica de Altura Vs PresiÃ³n (julio)
dfj1= dfj[['PRES', 'HGHT']] 
mpl.figure();dfj1.plot(x='PRES');mpl.legend(loc='best')
mpl.gca().invert_xaxis()
mpl.title("PresiÃ³n/Altura")
mpl.ylabel("Altura")
mpl.xlabel("PresiÃ³n")
mpl.grid(True)
mpl.show()
```
