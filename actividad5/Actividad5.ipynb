```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mpl
from datetime import datetime
```

```python
# Leer archivo de datos 00Z
# Convertir la columna CAPE de objeto a n√∫mero
df0 = pd.read_csv("00Z.csv", header=None, names=['Date', 'CAPE', 'PW'])
df0.CAPE=pd.to_numeric(df0.CAPE, errors='coerce')
df0.head()
```

```python
df0.dtypes
```

```python
# Convertir la cadena de caracteres 'Date' en variable temporal 'NDate'
df0['Ndate'] = pd.to_datetime(df0['Date'], format= '%d %m %Y')
df0['month'] = df0['Ndate'].dt.month
df0.head()
```

```python
# graficar Boxplots por mes 00Z
# Biblioteca Seaborn
import seaborn as seb
ax = seb.boxplot(x="month", y="CAPE", data=df0)
mpl.show()
```

```python
# graficar Boxplots por mes 00Z
# Biblioteca Seaborn
ax = seb.boxplot(x="month", y="PW", data=df0)
mpl.show()
```

```python
#Datos 00Z
seb.set(style="darkgrid", color_codes=True)

g = seb.jointplot("CAPE", "PW", data=df0, kind="reg",
                   color="y", size=7)
mpl.show(g)
```

```python
#Datos 00Z
g = seb.lmplot(x="CAPE", y="PW", hue="month",
               truncate=True, size=5, data=df0)
mpl.show(g)
```
