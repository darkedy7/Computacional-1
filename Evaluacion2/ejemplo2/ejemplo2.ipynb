```python
import numpy as np
from scipy.integrate import odeint
from mpl_toolkits.mplot3d import Axes3D

rho = 28.0
sigma = 10.0
beta = 8.0 / 3.0

def f(state, t):
    x, y, z = state  # unpack the state vector
    return sigma * (y - x), x * (rho - z) - y, x * y - beta * z  # derivatives

state0 = [0.1, 0.0, 0.0]
t = np.arange(0.0, 100.0, 0.01)

states = odeint(f, state0, t)
x = states[:, 0]
y = states[:, 1]
z = states[:, 2]
import matplotlib.pyplot as mpl
# plot the x values vs the y values
mpl.plot(t,x, color='r', alpha=0.7, linewidth=0.3)
mpl.title('x(t)')
mpl.savefig('Xt.png', dpi=180, bbox_inches='tight')
mpl.show()
#plot X Vs Z Valores
mpl.plot(t,y , color='m', alpha=0.7, linewidth=0.3)
mpl.title('y(t)')
mpl.savefig('Yt.png', dpi=180, bbox_inches='tight')
mpl.show()
# plot  Y  vs  Z 
mpl.plot(t,z, color='b', alpha=0.7, linewidth=0.3)
mpl.title('z(t)')
mpl.savefig('Zt.png', dpi=180, bbox_inches='tight')
mpl.show()
```
